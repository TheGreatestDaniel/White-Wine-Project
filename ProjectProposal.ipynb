## White Wine quality dataset

### this dataset contains mutiple factor columns that decide what people think good white wine is

### Fixed Acidity
Non-volatile acids like tartaric acid, providing wine with structure and crispness.

### Volatile Acidity
Evaporative acids like acetic acid; low levels add complexity, high levels cause spoilage.

### Citric Acid
Minor acid enhancing freshness and zesty flavors, often adjusted in winemaking.

### Residual Sugar
Unfermented sugar that defines sweetness, balancing acidity.

### Cholorides
 Salts influencing mouthfeel and perceived salinity in wine.

### Free Sulfur Dioxide
Active form of SO₂, protecting wine from spoilage and oxidation.

### Total Sulfur Dioxide
Combined free and bound SO₂ levels, crucial for wine preservation.

### Density
 Indicates sugar content and potential alcohol during fermentation.

### pH
Measures acidity strength, affecting taste, stability, and aging.

### Sulfates
Compounds enhancing structure and bitterness, often linked to minerality.

### Alcohol
 Ethanol content, contributing to body, warmth, and flavor balance.
